```python
import pandas as pd

# Function to load and preprocess each file
def load_and_preprocess(file_path, skip_rows):
    df = pd.read_csv(file_path, skiprows=skip_rows, low_memory=False)
    df['Glucose'] = df['Scan Glucose mmol/L'].combine_first(df['Historic Glucose mmol/L'])
    df['Glucose'] = pd.to_numeric(df['Glucose'], errors='coerce')
    df['Timestamp'] = pd.to_datetime(df['Device Timestamp'], dayfirst=True, errors='coerce')
    df = df[['Timestamp', 'Glucose']].dropna()
    df = df.set_index('Timestamp').sort_index()
    return df[['Glucose']]

# Load and preprocess each file
df_glucose1 = load_and_preprocess('BG.csv', skip_rows=2)
df_glucose2 = load_and_preprocess('BG-1.csv', skip_rows=2)

# Concatenate the two DataFrames
df_combined = pd.concat([df_glucose1, df_glucose2])

# Handle duplicates by averaging glucose values if they share the same timestamp

# Sort the combined data by timestamp
df_combined = df_combined.sort_index()

# Save the combined data to a CSV file without specific formatting

# Display the combined and sorted data
df_combined
```
