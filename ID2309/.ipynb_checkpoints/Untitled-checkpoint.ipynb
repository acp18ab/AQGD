```python
import pandas as pd
from datetime import datetime

def read_and_process(file_path):
    df = pd.read_csv(file_path)
    df['Timestamp'] = df.apply(lambda row: datetime.utcfromtimestamp(row['startTimeInSeconds'] + row['startTimeOffsetInSeconds']).strftime('%Y-%m-%d %H:%M:%S'),axis=1)
    df = df[['Timestamp', 'activityType', 'activeKilocalories', 'steps', 'distanceInMeters', 'durationInSeconds', 
             'activeTimeInSeconds', 'startTimeInSeconds', 'startTimeOffsetInSeconds', 
             'met', 'intensity', 'meanMotionIntensity', 'maxMotionIntensity']]
    return df

file_paths = [
    '06-02-2024---06-03-2024.csv',
    '06-03-2024---04-04-2024.csv',
    '04-04-2024---30-04-2024.csv'
]

df_list = [read_and_process(file_path) for file_path in file_paths]
df_activity = pd.concat(df_list)
df_activity['Timestamp'] = pd.to_datetime(df_activity['Timestamp'])
df_activity.set_index("Timestamp", inplace=True)
df_activity.sort_index(inplace=True)
df_activity.reset_index(inplace=True)
df_activity.to_csv("Activity_Epoch.csv", index = False)
df_activity
```
